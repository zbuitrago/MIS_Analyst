## Observaciones 
* se observa que las tablas para diferentes tablas independientemente del periodo o Update las columnas son iguales. los registros son actualizados con una periodicidad es esta misma con la se generan diferentes archivos segn tiempo estipulado para cada uno.

Esto se ve reflejado en las tablas de:
