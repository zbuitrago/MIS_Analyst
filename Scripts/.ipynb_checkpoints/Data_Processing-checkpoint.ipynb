```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import glob
```

## Poner PATH de archivo

```python
pathArchivos = input("Escriba el PATH directorio / carpeta de su archivo")
```

```python
# importar archivo de ubicacion
Tabla_Centro_Costo_df = pd.read_excel(pathArchivos + "Tabla Centros de Costo.xlsx")
Tabla_Areas_Comunes_df = pd.read_excel(pathArchivos + "Tabla areas comunes.xlsx")
```

```python
# Cargar de informacion Tabla de arriendos en un solo data dataframe 
allFiles = glob.glob(pathArchivos + "Tabla Arriendos Mes ?.xlsx")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_excel(file_,index_col=None, header=0)
    list_.append(df)
Tabla_Arriendos_df = pd.concat(list_)
Tabla_Arriendos_df['Update'] = pd.DatetimeIndex(Tabla_Arriendos_df['Update']).year
```

```python
# Cargar de informacion Tabla Areas Comunes Mes en un solo data dataframe 
allFiles = glob.glob(pathArchivos + "Tabla Areas Comunes Mes ?.xlsx")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_excel(file_,index_col=None, header=0)
    list_.append(df)
Tabla_Areas_Comunes_Mes_df = pd.concat(list_)
```

```python
# Cargar de informacion Tabla Seats Input Mes en un solo data dataframe 
allFiles = glob.glob(pathArchivos + "Tabla Seats Input Mes ?.xlsx")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    Tabla_Seats_Input_Mes_df= pd.read_excel(file_,index_col=None, header=0)
    list_.append(Tabla_Seats_Input_Mes_df)
Tabla_Seats_Input_Mes_df= pd.concat(list_)
Tabla_Seats_Input_Mes_df['Period'] = pd.to_datetime(Tabla_Seats_Input_Mes_df['Period']).dt.to_period('M')
Tabla_Seats_Input_Mes_df=Tabla_Seats_Input_Mes_df.rename(columns={'Period': 'Year-Month'})
Tabla_Seats_Input_Mes_df.head(1)
```
